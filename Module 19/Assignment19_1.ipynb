```python
### GRADED
uf = list()
# YOUR CODE HERE
for col in reviews.columns[:-2]:
  col_nn=reviews[reviews[col].isna()==False]
  X = col_nn[['F1','F2']]
  y = col_nn[col]
  lr = LinearRegression(fit_intercept=False).fit(X,y)
  coefs = lr.coef_
  uf.append(coefs)

uf =np.array(uf)

### ANSWER CHECK
uf.shape #should be (5, 2)
```
